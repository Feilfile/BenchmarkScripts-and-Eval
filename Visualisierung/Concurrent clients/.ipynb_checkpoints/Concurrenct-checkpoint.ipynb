```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_and_prepare_data(file_path):
    df = pd.read_csv(file_path)
    df['Cluster Setup'] = df['Amount of Nodes'].astype(str) + ' x ' + df['Size per Node'].astype(str) + ', ' + df['Amount of CClients'].astype(str) + ' Clients'
    return df

def plot_ops_by_clients_for_setups(df, setups, color_palette):
    for setup in setups:
        plt.figure(figsize=(10, 6))

        df_filtered = df[df['Cluster Setup'].str.startswith(setup)]

        sns.barplot(x='Amount of CClients', y='Ops/sec', hue='Benchmark Method', data=df_filtered, palette=color_palette)
        plt.title(f'Ops/sec by Concurrent Clients - {setup} Setup')
        plt.ylabel('Operations per Second (Ops/sec)')
        plt.xlabel('Amount of Concurrent Clients')
        plt.legend(title='Benchmark Method')
        plt.show()

file_sets = './sets.csv'  # Replace with the path to your sets.csv file
file_gets = './gets.csv'  # Replace with the path to your gets.csv file

df_sets = load_and_prepare_data(file_sets)
df_gets = load_and_prepare_data(file_gets)

cluster_setups = ['1 x 30', '3 x 10', '5 x 6']

color_palette = {'memcached': 'blue', 'redis': 'green', 'redisAggregated': 'red'}

plot_ops_by_clients_for_setups(df_sets, cluster_setups, color_palette)
plot_ops_by_clients_for_setups(df_gets, cluster_setups, color_palette)
```
