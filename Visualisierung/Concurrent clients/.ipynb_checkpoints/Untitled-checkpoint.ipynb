```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load data
def load_data(file_path):
    return pd.read_csv(file_path)

# Function to plot data
def plot_benchmark_data(df, title):
    plt.figure(figsize=(12, 6))
    
    # Plotting Ops/sec
    sns.barplot(x='Benchmark Method', y='Ops/sec', hue='Amount of CClients', data=df)
    plt.title(title + " - Operations per Second")
    plt.ylabel('Operations per Second (Ops/sec)')
    plt.xlabel('Benchmark Method')
    plt.xticks(rotation=45)
    plt.legend(title='Concurrent Clients')
    plt.tight_layout()
    plt.show()

    # Plotting Average Latency
    plt.figure(figsize=(12, 6))
    sns.barplot(x='Benchmark Method', y='Avg. Latency', hue='Amount of CClients', data=df)
    plt.title(title + " - Average Latency")
    plt.ylabel('Average Latency (ms)')
    plt.xlabel('Benchmark Method')
    plt.xticks(rotation=45)
    plt.legend(title='Concurrent Clients')
    plt.tight_layout()
    plt.show()

# Load the CSV files
file_sets = './sets.csv'  # Replace with your file path
file_gets = './gets.csv'  # Replace with your file path

# Reading the data
df_sets = load_data(file_sets)
df_gets = load_data(file_gets)

# Plotting for Sets and Gets Data
plot_benchmark_data(df_sets, 'Sets Benchmark')
plot_benchmark_data(df_gets, 'Gets Benchmark')

# Creating tables for Ops/sec
ops_sec_sets = df_sets[['Benchmark Method', 'Amount of Nodes', 'Amount of CClients', 'Ops/sec']]
ops_sec_gets = df_gets[['Benchmark Method', 'Amount of Nodes', 'Amount of CClients', 'Ops/sec']]

print("Operations per Second (Sets):")
print(ops_sec_sets)

print("\nOperations per Second (Gets):")
print(ops_sec_gets)
```

```python
def load_and_prepare_data(file_path):
    df = pd.read_csv(file_path)
    df['Cluster Setup'] = df['Amount of Nodes'].astype(str) + ' x ' + df['Size per Node'].astype(str) + ', ' + df['Amount of CClients'].astype(str) + ' Clients'
    return df

# Function to plot detailed comparison for specific cluster setups
def plot_specific_cluster_setups(df, setups, colors):
    for setup in setups:
        plt.figure(figsize=(10, 6))

        # Filtering the dataframe for the specific cluster setup
        df_filtered = df[df['Cluster Setup'] == setup]

        # Creating the plot
        sns.barplot(x='Benchmark Method', y='Ops/sec', data=df_filtered, palette=colors)
        plt.title(f'Ops/sec Comparison - {setup}')
        plt.ylabel('Operations per Second (Ops/sec)')
        plt.xlabel('Benchmark Method')
        plt.xticks(rotation=45)
        plt.show()

# Loading the data
df_sets = load_and_prepare_data(file_sets)
df_gets = load_and_prepare_data(file_gets)

# Defining the specific cluster setups and colors
cluster_setups = ['1 x 30, 1 Clients', '1 x 30, 5 Clients', '3 x 10, 1 Clients', '5 x 6, 1 Clients']
colors = ['red', 'blue', 'yellow']

# Plotting for Sets and Gets Data
plot_specific_cluster_setups(df_sets, cluster_setups, colors)
plot_specific_cluster_setups(df_gets, cluster_setups, colors)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Function to load and prepare data
def load_and_prepare_data(file_path):
    df = pd.read_csv(file_path)
    df['Cluster Setup'] = df['Amount of Nodes'].astype(str) + ' x ' + df['Size per Node'].astype(str) + ', ' + df['Amount of CClients'].astype(str) + ' Clients'
    return df

# Function to plot Ops/sec for each benchmark group by the number of concurrent clients for specific cluster setups
def plot_ops_by_clients_for_setups(df, setups, colors):
    for setup in setups:
        plt.figure(figsize=(10, 6))

        # Filtering the dataframe for the specific cluster setup
        df_filtered = df[df['Cluster Setup'].str.startswith(setup)]

        # Creating the plot
        sns.barplot(x='Amount of CClients', y='Ops/sec', hue='Benchmark Method', data=df_filtered, palette=colors)
        plt.title(f'Ops/sec by Concurrent Clients - {setup} Setup')
        plt.ylabel('Operations per Second (Ops/sec)')
        plt.xlabel('Amount of Concurrent Clients')
        plt.legend(title='Benchmark Method')
        plt.show()

# Loading the data
df_sets = load_and_prepare_data(file_sets)
df_gets = load_and_prepare_data(file_gets)

# Defining the specific cluster setups and colors
cluster_setups = ['1 x 30', '3 x 10', '5 x 6']
colors = ['red', 'purple', 'green']  # Assign colors in the order of the benchmark methods in your data

# Plotting for Sets and Gets Data
plot_ops_by_clients_for_setups(df_sets, cluster_setups, colors)
plot_ops_by_clients_for_setups(df_gets, cluster_setups, colors)
```
