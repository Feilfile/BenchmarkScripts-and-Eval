```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_and_prepare_data_with_datasize(file_path):
    df = pd.read_csv(file_path)
    df['Cluster Setup'] = df['Amount of Nodes'].astype(str) + ' x ' + df['Size per Node'].astype(str)
    return df

def plot_ops_by_datasize_for_setups(df, setups, colors):
    for setup in setups:
        plt.figure(figsize=(10, 6))

        df_filtered = df[df['Cluster Setup'] == setup]

        sns.barplot(x='Data Size (Bytes)', y='Ops/sec', hue='Benchmark Method', data=df_filtered, palette=colors)
        plt.title(f'Ops/sec by Data Size - {setup} Setup')
        plt.ylabel('Operations per Second (Ops/sec)')
        plt.xlabel('Data Size (Bytes)')
        plt.legend(title='Benchmark Method')
        plt.show()

file_sets = './sets.csv'  
file_gets = './gets.csv' 

df_sets_with_datasize = load_and_prepare_data_with_datasize(file_sets)
df_gets_with_datasize = load_and_prepare_data_with_datasize(file_gets)

cluster_setups = ['1 x 30', '3 x 10', '5 x 6']

color_palette = {'memcached': 'blue', 'redis': 'green', 'redisAggregated': 'red'}

plot_ops_by_datasize_for_setups(df_sets_with_datasize, cluster_setups, color_palette)
plot_ops_by_datasize_for_setups(df_gets_with_datasize, cluster_setups, color_palette)
```
